```python
import pandas as pd
from category_encoders import HashingEncoder
```

```python
df = pd.read_csv('../datos/procesados/adult_limpio.csv')
```

```python
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
```

```python
df.info()
```

# ONE HOT ENCODING

```python
df_ohe = pd.get_dummies(df, dtype=int)
```

```python
df_ohe.head()
```

```python
df_ohe.info()
```

```python
df_ohe.columns.tolist()
```

```python
df_ohe.to_csv('../datos/procesados/adult_ohe.csv', index=False)
```

# COUNT ENCODING

```python
df_count = df.copy()
```

```python
cat_cols = df_count.select_dtypes(include=['object']).columns

for col in cat_cols:
    freq = df_count[col].value_counts()
    df_count[col] = df_count[col].map(freq)
```

```python
df_count.head()
```

```python
df_count.info()
```

```python
df_count.to_csv('../datos/procesados/adult_count.csv')
```

# HASH ENCODING

```python
col_categoricas = ['workclass', 'education', 'marital-status', "occupation", "relationship", "race", "gender", "native-country"]
df_hash_8 = HashingEncoder(cols=col_categoricas, n_components=8).fit_transform(df)
df_hash_16 = HashingEncoder(cols=col_categoricas, n_components=16).fit_transform(df)
df_hash_32 = HashingEncoder(cols=col_categoricas, n_components=16).fit_transform(df)
```
